```python
import google.generativeai as genai

genai.configure(api_key="Enter Gemini API Key")

model = genai.GenerativeModel("gemini-3.1-flash-lite")

user_text = input("Paste your text: ")

def summarize_text(text):

    prompt = f"""
    You are an expert note-taking assistant.

    Analyze the following text and provide:

    1. Short Summary
    2. Key Points
    3. Important Action Items
    4. Simplified Explanation

    Text:
    {text}
    """

    response = model.generate_content(prompt)

    return response.text

result = summarize_text(user_text)

print("🧠 SMART NOTES GENERATOR")
print("-------------------------")
print(result)
print("-------------------------")
print("✅ Notes generated successfully!")
```
